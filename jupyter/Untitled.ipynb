## Asymptotic Complexity
There is a lot of material talking about the asymptotic complexities, and this is nothing new. I just wanted to put a assemeble the comparisons between different algorithms at one place

```python
%pylab --no-import-all
%matplotlib inline

fig,ax = plt.subplots()

x = range(100)
yfact = [1]
ysnom = [1]
ysq = [0]
ysuplog = [0]
ylin = [0]
ylog = [0]
for idx in xrange(1, len(x)):
    yfact.append(yfact[-1] * idx)
    ysnom.append(2**idx)
    ysq.append(idx**2)
    ysuplog.append(idx*np.log2(idx))
    ylin.append(idx)
    ylog.append(np.log2(idx))
plt.plot(x,yfact,linewidth=2,label='$O(x!)$')
plt.plot(x,ysnom,linewidth=2,label='$O(2^x)$')
plt.plot(x,ysq,linewidth=2,label='$O(x^2)$')
plt.plot(x,ysuplog,linewidth=2,label='$O(x\log{x})$')
plt.plot(x,ylin,linewidth=2,label='$O(x)$')
plt.plot(x,ylog,linewidth=2,label='$O(\log{x})$')
plt.axis([0,50, 0, 100])
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.legend()
plt.title('Asymptotic Complexity')
```
